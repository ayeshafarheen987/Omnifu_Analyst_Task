```python

# Group by 'Week' and 'Month'
weekly_data = df.groupby('Week').sum().reset_index()
monthly_data = df.groupby('Month').sum().reset_index()

# Calculate Returns % (Leads divided by Clicks)
weekly_data['Returns %'] = (weekly_data['Leads'] / weekly_data['Clicks']) * 100
monthly_data['Returns %'] = (monthly_data['Leads'] / monthly_data['Clicks']) * 100

# Display the weekly report
print("Weekly Report:")
print(weekly_data[['Week', 'Clicks', 'Impressions', 'Cost ($)', 'Leads', 'Prospects', 'Returns %']])

# Display the monthly report
print("\nMonthly Report:")
print(monthly_data[['Month', 'Clicks', 'Impressions', 'Cost ($)', 'Leads', 'Prospects', 'Returns %']])
```

```python
import matplotlib.pyplot as plt

# Convert 'Month' column to string format
monthly_data['Month'] = monthly_data['Month'].dt.strftime('%Y-%m')

# Line plot for weekly data
plt.figure(figsize=(10, 6))
plt.plot(weekly_data['Week'], weekly_data['Clicks'], label='Clicks')
plt.plot(weekly_data['Week'], weekly_data['Leads'], label='Leads')
plt.plot(weekly_data['Week'], weekly_data['Cost ($)'], label='Cost ($)')
plt.xlabel('Week')
plt.ylabel('Metrics')
plt.title('Weekly Metrics')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Line plot for monthly data
plt.figure(figsize=(10, 6))
plt.plot(monthly_data['Month'], monthly_data['Clicks'], label='Clicks')
plt.plot(monthly_data['Month'], monthly_data['Leads'], label='Leads')
plt.plot(monthly_data['Month'], monthly_data['Cost ($)'], label='Cost ($)')
plt.xlabel('Month')
plt.ylabel('Metrics')
plt.title('Monthly Metrics')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
